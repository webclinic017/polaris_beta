Import libraries and instantiate objects.

```python
from polaristools.binanceconnection import BinanceConnection
import pandas as pd
```

```python
binance = BinanceConnection()
```

# __ SPOT MARGIN MARKET __
## The greatest traded volumes in 24 hours, quoted in USDT.

```python
rollingStats = binance.tickerPriceChangeStatistics24hRoll()
```

```python
df = pd.DataFrame(rollingStats)
pd.options.display.float_format = "{:.4f}".format
```

```python
usdt_quoted = df[df['symbol'].str.endswith('USDT')==True].copy()
usdt_quoted = usdt_quoted[['symbol','priceChange','priceChangePercent','weightedAvgPrice','volume','quoteVolume']].copy()


tofloat64 = ['priceChange','priceChangePercent','weightedAvgPrice','volume','quoteVolume']
usdt_quoted[tofloat64] = usdt_quoted[tofloat64].astype('float64')


usdt_quoted.sort_values(
    by='quoteVolume', 
    ascending=False,
    inplace=True,
)
```

```python
stable_coins = ['BUSDUSDT','USDCUSDT','EURUSDT']
others_rejected = []
rejected_symbols = stable_coins + others_rejected
usdt_quoted = usdt_quoted[~usdt_quoted['symbol'].isin(rejected_symbols)].copy()
```

```python
usdt_quoted_30 = usdt_quoted['symbol'][0:30]
```

# __ FUTURES USD MARKET __
## The greatest traded volumes in 24 hours, quoted in USDT.

```python
rollingStats_futures = binance.futuresTickerPriceChangeStatistics24h()
```

```python
df = pd.DataFrame(rollingStats_futures)
pd.options.display.float_format = "{:.4f}".format
usdt_quoted = df[
                    [
                    'symbol',
                    'priceChange',
                    'priceChangePercent',
                    'weightedAvgPrice',
                    'volume',
                    'quoteVolume',
                    ]
                ][df['symbol'].str.endswith('USDT')==True].copy()

tofloat64 = ['priceChange','priceChangePercent','weightedAvgPrice','volume','quoteVolume']
usdt_quoted[tofloat64] = usdt_quoted[tofloat64].astype('float64')

usdt_quoted.sort_values(
                        by='quoteVolume', 
                        ascending=False,
                        inplace=True,
                        )

stable_coins = ['BUSDUSDT','USDCUSDT','EURUSDT']
others_rejected = []
rejected_symbols = stable_coins + others_rejected
usdt_quoted_fut = usdt_quoted[~usdt_quoted['symbol'].isin(rejected_symbols)].copy()
```

```python
usdt_quoted_fut.nlargest(30, 'quoteVolume')
```
